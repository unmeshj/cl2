```python
#step 1 -- import libraries 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#step 2 -- generate data using make_blobs
X,y=make_blobs(n_samples=100,centers=3,cluster_std=1.05,random_state=42)
```

```python
# X,y
```

```python
# step3 -- 3. Visualize Data
plt.scatter(X[:,0],X[:,1],c='blue',marker='o',edgecolor='black',s=50)
plt.title("Generated Data")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

```python
# Step 4 --> Apply Agglomerative Cust
AgglomerativeClustering=AgglomerativeClustering(n_clusters=3, metric='euclidean',linkage='ward')#affinity='euclidean'
labels=AgglomerativeClustering.fit_predict(X)
```

```python
# Step 5 --> visualize Clustering result
# Plot clustering results
plt.scatter(X[:, 0], X[:, 1], c=labels, marker='o', edgecolor='black', s=50, cmap='viridis')
plt.title("Agglomerative Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
#Step 6(Optional) Compute the linkage matrix
linked = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Dendrogram")
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```
